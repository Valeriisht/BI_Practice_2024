```python
# import

import matplotlib.pyplot as plt
import pandas as pd
import os
import seaborn as sns
import numpy as np


```

```python
dir = "Project3/results/jellyfish/"

# Загрузка данных из файла
data = pd.read_csv(dir+'PE_histo.txt', sep=' ')
data = data.loc[:, ["14151553"]]
```

```python
# переименовывания таблицы
data.columns = ["k-mer 31 frequency"]
```

```python
#  распределение частоты встречаемости
sns.lineplot(data)
plt.title('Frequency distribution')
plt.xlim(0, 200)
plt.ylim(0, 200000)
plt.xlabel("Coverage")
plt.ylabel("Frequency")
plt.savefig("histo_jellyfish",  bbox_inches="tight", dpi=500)
plt.show()
```
